Q4:- Describe how dictionaries store data.

Ans4:- A dictionary stores data in the form of key–value pairs.  
Internally, dictionaries use a hash table. When a key is added, Python applies a hash function to the key to generate a unique hash value. This hash value determines the location in memory where the corresponding value is stored. Because of this hashing mechanism, dictionaries provide very fast data access, usually in constant time O(1).

Q8:- What is a hash table, and how does it relate to dictionaries in Python?


Ans8:- A hash table is a data structure that stores data as key–value pairs and uses a hash function to compute an index (called a hash) into an array, where the value is stored.  
### How Hash Tables Relate to Python Dictionaries
#### In Python, dictionaries are implemented using hash tables.  
When you store a key-value pair in a dictionary, Python:  
#### 1. Applies a hash function to the key.  
#### 2. Finds the location in memory based on the hash.  
#### 3. Stores the value at that location.

Q17:- Describe the time complexity of accessing elements in a dictionary.


Ans17:-Accessing elements in a dictionary by key is O(1) on average because dictionaries use hash tables, making them extremely efficient for lookups.
A dictionary in Python stores data as key–value pairs using a hash table. This design allows Python to access values directly using their keys, without having to search through all the elements.  

    Average Case: Accessing a value by key is O(1) (constant time). Python computes the hash of the key, which gives the memory location where the   value is stored, allowing immediate retrieval.

     Worst Case: In rare situations, such as hash collisions, access can degrade to O(n), where n is the number of elements. However, Python’s hash table implementation minimizes this, so such cases are uncommon.

Q19:- Why are dictionaries considered unordered, and how does that affect data retrieval?

Ans19:- Dictionaries are considered unordered because their internal storage is hash-based, meaning keys are not stored in a sequence. This affects data retrieval by requiring access always via keys rather than positions, but it enables fast lookup and efficient storage.   

       Indexing Not Supported

     You cannot access elements by index like a list.  
     
     Lookup speed    Very fast (O(1))
